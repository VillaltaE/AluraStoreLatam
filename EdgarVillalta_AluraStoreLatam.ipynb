### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#Se estima el ingreso de cada tienda sumando todos los Precios
ingreso_tienda = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()


#Se estima el ingreso de cada tienda sumando todos los Precios
ingreso_tienda = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

print(f'El ingreso de tienda 1 en millones de pesos  : {float(ingreso_tienda/1000000):.2f}')
print(f'El ingreso de tienda 2  en millones de pesos es : {float(ingreso_tienda2/1000000):.2f}')
print(f'El ingreso de tienda 3 en millones de pesos es : {float(ingreso_tienda3/1000000):.2f}')
print(f'El ingreso de tienda 4 e en millones de pesos es : {float(ingreso_tienda4/1000000):.2f}')
# Se suman los ingresos de cada tienda
ingresos_totales = [ingreso_tienda, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

#Para obtener el ingreso total se suman los ingresos de cada tienda
ingreso_total = sum(ingresos_totales)

print(f'El ingreso total en millones de pesos  es : {float(ingreso_total/1000000):.2f}')
#Tienda con mayor ingreso
mayor_ingreso = max(ingresos_totales)


#Tienda con menor ingreso
menor_ingreso = min(ingresos_totales)

#Tienda es la de mayor ingreso
print(f'El mayor ingreso en millones de pesos  es :, {float(mayor_ingreso/1000000):.2f} tienda 1')
#Tienda4 es la de menor ingreso
print(f'El menor ingreso en millones de pesos es : {float(menor_ingreso/1000000):.2f} tienda 4')
```

Explore si algunas tiendas tienen un rendimiento superior o inferior al esperado en determinadas regiones e identifique si existen patrones geográficos que puedan influir en el rendimiento de las tiendas.

```python
# Estimacion de ventas por par de coordenadas
n_ventas = tienda[['lat','lon']].value_counts()

# Se prepara una data frame con solo coordenas
coor_tienda = tienda[['lat','lon']].drop_duplicates()

#Se agrega las ventas  a cada par de coordenadas
coor_tienda['Ventas'] = coor_tienda.apply(lambda row: n_ventas.get((row['lat'], row['lon']), 0), axis=1) # Use apply and get to assign sales counts to coor_tienda3
print(coor_tienda)
```
